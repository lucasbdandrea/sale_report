```python
import pandas as pd
```

```python
df = pd.read_excel('../data/processed_data/Meganium_Sales_data.xlsx')
df.head()
```

```python
df.info()
```

```python
# Analisando a quantidade de vendas de cada produto por pais
qtd_por_pais = df.groupby(['delivery_country', 'product_sold'])['quantity'].sum().reset_index().sort_values(
    ['delivery_country', 'quantity'], ascending=[True, False]
).reset_index(drop=True)
qtd_por_pais
```

```python
# Selecionando apenas o produto mais vendido por pais (valor)
maiores_vendas_qtd = qtd_por_pais.groupby('delivery_country').first().reset_index()
maiores_vendas_qtd
```

```python
# Analisando o valor arrecadado por cada produto por pais s/cupom
valor_por_pais = df.groupby(['delivery_country', 'product_sold'])['total_price'].sum().reset_index().sort_values(
    ['delivery_country', 'total_price'], ascending=[True, False]
).reset_index(drop=True)
valor_por_pais
```

```python
# Selecionando apenas o produto mais vendido por pais (valor)
maiores_vendas_valor = valor_por_pais.groupby('delivery_country').first().reset_index()
maiores_vendas_valor
```

```python
# Analisando o valor de desconto por produto em cada pais
desconto_por_pais = df.groupby(['delivery_country', 'product_sold'])[
    'discount_value'].sum().reset_index().sort_values(
    ['delivery_country', 'discount_value'], ascending=[True, False]
).reset_index(drop=True)
desconto_por_pais
```

```python
# Selecionando apenas os produtos com mais desconto por pais 
maiores_descontos = desconto_por_pais.groupby('delivery_country').first().reset_index()
maiores_descontos
```

```python
#verificando se o produto mais vendido é o que mais arrecadou
maiores_vendas_qtd['product_sold'] == maiores_vendas_valor['product_sold']
```

```python
#verificando se o mais vendido é o com o maior desconto
maiores_vendas_qtd['product_sold'] == maiores_descontos['product_sold']
```

```python
from datetime import datetime

df['age'] = df['buyer_birth_date'].apply(
    lambda x: datetime.now().year - x.year if pd.notnull(x) else None)

#verificando a faixa etária média dos clientes por produto e pais
faixa_etaria = df.groupby(['delivery_country', 'product_sold'])['age'].mean().reset_index()
faixa_etaria.merge(
    desconto_por_pais, on=['delivery_country', 'product_sold'], how='inner')
```

```python
faixa_etaria = faixa_etaria.merge(
    maiores_vendas_qtd, on=['delivery_country', 'product_sold'], how='inner')
faixa_etaria
```
